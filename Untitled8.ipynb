```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
class UrbanDevelopmentDataset(Dataset):
    def __init__(self, split='train', image_size=256, target_size=64):
        self.dataset = load_dataset("blanchon/LEVIR_CDPlus", split=split)
        self.image_size = image_size  # Input size
        self.target_size = target_size  # Model output size

        # Image transformation
        self.image_transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Resize((self.image_size, self.image_size)),  # Resize input
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        # Mask transformation (resize to model's output size)
        self.mask_transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Resize((self.target_size, self.target_size), transforms.InterpolationMode.NEAREST)  # Match model output
        ])

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]

        # Load images
        image_before = np.array(item['image1'])  # Use 'image1' as input
        mask = np.array(item['mask'])  # Load segmentation mask

        # Convert & resize
        image_tensor = self.image_transform(image_before)
        mask_tensor = self.mask_transform(mask).float()  # Ensure mask is float

        return image_tensor, mask_tensor
```

```python
# Initialize datasets
train_dataset = UrbanDevelopmentDataset('train')
val_dataset = UrbanDevelopmentDataset('test')

# Create data loaders
BATCH_SIZE = 8
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
```

```python
class UrbanPlanningUNet(nn.Module):
    def __init__(self):
        super().__init__()

        # Encoder (pretrained ResNet-18)
        self.encoder = torch.hub.load('pytorch/vision', 'resnet18', pretrained=True)
        self.encoder_layers = list(self.encoder.children())[:-2]

        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 256, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 1, 1)
        )

    def forward(self, x):
        # Encoder
        for layer in self.encoder_layers:
            x = layer(x)

        # Decoder
        x = self.decoder(x)
        return torch.sigmoid(x)

model = UrbanPlanningUNet()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
NUM_EPOCHS = 2
```

```python
for epoch in range(NUM_EPOCHS):
    model.train()
    train_loss = 0

    for images, masks in train_loader:
        images = images.to(device)
        masks = masks.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()

        train_loss += loss.item() * images.size(0)

    # Validation
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for images, masks in val_loader:
            images = images.to(device)
            masks = masks.to(device)
            outputs = model(images)
            loss = criterion(outputs, masks)
            val_loss += loss.item() * images.size(0)

    print(f"Epoch {epoch+1}/{NUM_EPOCHS}")
    print(f"Train Loss: {train_loss/len(train_dataset):.4f}")
    print(f"Val Loss: {val_loss/len(val_dataset):.4f}\n")
```

```python
def visualize_development(image_path='/content/sample_data/pleiades-1-airplane-graveyard.jpeg',
                         model=None,
                         threshold=0.5,
                         device='cpu'):  # Added device argument
    """Visualizes optimal development sites using a segmentation model."""

    try:
        # 1. Image Loading and Preprocessing
        try:
            image = Image.open(image_path).convert('RGB')
            original = np.array(image)
            print(f"Image loaded successfully from: {image_path}, shape: {original.shape}")
        except FileNotFoundError:
            print(f"Error: Image not found at {image_path}")
            return
        except Exception as e:
            print(f"Error loading image: {e}")
            return

        # Resize if necessary
        if original.shape[:2] != (256, 256):
            image = image.resize((256, 256))
            original = np.array(image)  # Update original after resizing
            print("Image resized to (256, 256)")

        # Normalization pipeline
        preprocess = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])
        preprocessed = preprocess(image)
        print("Image preprocessed successfully")

        # 2. Model Prediction
        try:
            model.eval()  # Set the model to evaluation mode
            with torch.no_grad():
                input_tensor = preprocessed.unsqueeze(0).to(device)  # Add batch dimension
                output = model(input_tensor)
                output = torch.sigmoid(output)  # Apply sigmoid if not already applied in the model
                mask = (output.squeeze().cpu().numpy() > threshold).astype(np.uint8)
            print(f"Model prediction successful, mask shape: {mask.shape}, unique values: {np.unique(mask)}")
        except Exception as e:
            print(f"Error during model prediction: {e}")
            return

        # 3. Visualization Setup
        try:
            overlay = cv2.resize(original.copy(), (256, 256))
            resized_mask = cv2.resize(mask.astype(np.uint8), (256, 256), interpolation=cv2.INTER_NEAREST) # Resize the mask
            overlay[resized_mask == 1] = [255, 0, 0]  # Red overlay for development zones
            print("Overlay created successfully")
        except Exception as e:
            print(f"Error creating overlay: {e}")
            return

        # 4. Plot Configuration
        try:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

            ax1.imshow(original)
            ax1.set_title('Input Satellite Imagery')
            ax1.axis('off')

            ax2.imshow(overlay)
            ax2.set_title('Optimal Development Sites')
            ax2.legend(handles=[plt.Line2D([0], [0], marker='s', color='w',
                                         markerfacecolor='red', markersize=10,
                                         label='Recommended Construction')],
                     loc='lower left')
            ax2.axis('off')

            plt.tight_layout()
            plt.show()
            print("Visualization plotted successfully")

        except Exception as e:
            print(f"Error during plotting: {e}")
            return

    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example Usage (Ensure 'model' is defined and trained):
# Assuming 'model' and 'device' are already defined in your notebook
visualize_development(image_path='/content/sample_data/pleiades-1-airplane-graveyard.jpeg',
                        model=model,
                        threshold=0.5,
                        device=device)
```
